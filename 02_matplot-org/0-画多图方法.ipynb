```python

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
%matplotlib inline


x = np.linspace(0,49,50)
y = np.sin(x)

fig = plt.figure()
#唐老师方法
ax1 = plt.subplot(211)
plt.plot(x,y)

ax2 = plt.subplot(212)
plt.plot(x,y/2)

plt.show()
```

```python
#官网
fig,(ax1, ax2) = plt.subplots(nrows=2)

ax1.plot(x,y)
ax2.plot(x,y/2)
```
